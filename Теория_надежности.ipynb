```python
class SerialConnection:
  def __init__(self, elements):
    self.elements = elements

  def count(self, t):
    p = 1
    for element in self.elements:
      p *= element.count(t)
    return p
```

```python
class TestElement:
  def __init__(self, p):
    self.p = p

  def count(self, t):
    return self.p
```

```python
class ParallelConnection:
  def __init__(self, elements):
    self.elements = elements
  
  def count(self, t):
    p = 1
    for element in self.elements:
      p *= (1 - element.count(t))
    return 1 - p
```

```python
from math import exp

class ExpElement:
  def __init__(self, h):
    self.h = h

  def count(self, t):
    return exp(-self.h * t)
```

```python
class WeibullElement:
  def __init__(self, a, b):
    self.a = a
    self.b = b
  
  def count(self, t):
    return exp(-((t/self.a)**self.b))
```
